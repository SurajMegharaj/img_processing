```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
class image_processing:
  def __init__(self,image_path):
    self.image = image_path
    self.kernel = np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]])
    if self.image == None:
      raise FileNotFoundError('No Image Found')
    else:
      self.img = cv2.imread(self.image)
      cv2_imshow(self.img)
      cv2.waitKey(0)
      cv2.destroyAllWindows()

  def resize_img(self,width,height):
    resize_img = cv2.resize(self.img,(width,height))
    cv2_imshow(resize_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return resize_img

  def grey_scale(self):
    grat_img = cv2.cvtColor(self.img,cv2.COLOR_BGR2GRAY)
    cv2_imshow(grat_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return grat_img

  def blur(self,ksize):
    blur_img = cv2.GaussianBlur(self.img,(ksize ,ksize),0)
    cv2_imshow(blur_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return blur_img

  def edge_detection(self):
    edge_img = cv2.Canny(self.img, 100,200)
    cv2_imshow(edge_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return edge_img

  def sharpen(self,resize_img):
    sharpen_img = cv2.filter2D(resize_img, -1, self.kernel)
    cv2_imshow(sharpen_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return sharpen_img
```

Resize image

```python
imm = image_processing('/content/images (3).jpg')
re = imm.resize_img(300,300)
imm.grey_scale()
imm.blur(1)
imm.edge_detection()
imm.sharpen(re)
```
